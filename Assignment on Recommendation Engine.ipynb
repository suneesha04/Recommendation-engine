### Euclidean Distance Metric for Recommendation system

```python
from sklearn.metrics import pairwise_distances
```

```python
books_new_user_euclidean = books_new_user.T
books_new_user_euclidean
```

```python
books_new_user_euclidean.values
```

```python
user_to_user = pairwise_distances(X = books_new_user_euclidean.values, metric='euclidean')
```

```python
user_to_user
```

```python
user_to_user_df = pd.DataFrame(data=user_to_user, columns = books_new['User.ID'].unique(), index = books_new['User.ID'].unique())
```

```python
user_to_user_df
```

### 3. using cosine Similarity for Recommand Engine

```python
user_to_user_cosine = 1- pairwise_distances(X = books_new_user_euclidean.values , metric = 'cosine')
```
